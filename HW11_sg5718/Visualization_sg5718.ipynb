```python
wcny = pd.read_csv("%s/WaterConsumption.csv"%os.getenv("PUIDATA"))
wcny.head()
```

```python
wcny.rename(columns = {'year':'Year', 
               'new_york_city_population':'NYC Population', 
               'per_capita_gallons_per_person_per_day': 'Water Consumption Per Capita(gallons)'}, inplace=True)
```

```python
wcny = wcny[['Year', 'Water Consumption Per Capita(gallons)', 'NYC Population']]
wcny.head()
```

```python
wcny.describe()
```

```python
wcny.sort_values(by='Year', inplace=True)

wcny.plot(x="Year", y="Water Consumption Per Capita(gallons)", figsize=(15, 10))
plt.title("Changes of Water Consumption per Person per Day in NYC through 1979 to 2017", fontsize=16)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Water Consumption per capita(gallons)", fontsize=20)
plt.yticks(np.arange(115, 220, 10.0))
plt.xticks(np.arange(1979, 2018, 2.0))
plt.show()
```

### Obviously, in general, water consumption per capita decreased during the period between 1979 and 2017.

```python
waterconsump = np.array(wcny["Water Consumption Per Capita(gallons)"])
y = waterconsump.tolist()
year = np.array(wcny["Year"])
x = year.tolist()
```

```python
figure = plt.figure(figsize=(15, 8))
wcnybar = plt.bar(range(len(y)), y,tick_label=x)
plt.title("Changes of Water Consumption per Person per Day in NYC through 1979 to 2017", fontsize=20)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Water Consumption per capita(gallons)", fontsize=20)
for r in wcnybar:
    plt.text(r.get_x() + r.get_width()/100, 1.004*r.get_height(), '%d' % int(r.get_height()))
plt.ylim(100,220)
plt.xticks(rotation=45)
plt.show()
```

### From the above histogram, we can see the water comsumption had a general decreasing trend through 1970 to 2017.

```python
wcny.sort_values(by='Year', inplace=True)

figure = plt.figure(figsize=(15, 8))
plt.scatter(x=wcny["Year"], y=wcny["NYC Population"])
plt.title("Changes of Population in NYC through 1979 to 2017", fontsize=20)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Population", fontsize=20)
plt.show()
```

### From the above plot,  we can see the population in New York had a increasing trend during the same period, which suggests that when population increased, the water consumption per capita decreases.
